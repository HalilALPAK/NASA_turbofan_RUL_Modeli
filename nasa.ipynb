```python
import pandas as pd

# Define column names based on the readme.txt
columns = ['unit_number', 'time_in_cycles', 'operational_setting_1', 'operational_setting_2', 'operational_setting_3'] + \
          ['sensor_measurement_' + str(i) for i in range(1, 22)]

# Load the training data
df_train = pd.read_csv('train_FD001.txt', sep='\s+', header=None, names=columns)

# Load the test data
df_test = pd.read_csv('test_FD001.txt', sep='\s+', header=None, names=columns)

# Display the head and info of the training dataframe
print("Training Data Head:")
display(df_train.head())
print("\nTraining Data Info:")
df_train.info()

# Display the head and info of the test dataframe
print("\nTest Data Head:")
display(df_test.head())
print("\nTest Data Info:")
df_test.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Drop the specified columns
drop_columns = ['operational_setting_3', 'sensor_measurement_1', 'sensor_measurement_5', 'sensor_measurement_10', 'sensor_measurement_16', 'sensor_measurement_18', 'sensor_measurement_19']
df_train_processed = df_train.drop(columns=drop_columns)
df_test_processed = df_test.drop(columns=drop_columns)

# Separate unit_number and time_in_cycles
train_unit_time = df_train_processed[['unit_number', 'time_in_cycles']]
test_unit_time = df_test_processed[['unit_number', 'time_in_cycles']]

df_train_processed = df_train_processed.drop(columns=['unit_number', 'time_in_cycles'])
df_test_processed = df_test_processed.drop(columns=['unit_number', 'time_in_cycles'])


# Normalize the remaining numerical features
scaler = MinMaxScaler()
df_train_scaled = scaler.fit_transform(df_train_processed)
df_test_scaled = scaler.transform(df_test_processed)

# Convert scaled data back to DataFrame to easily group by unit_number
df_train_scaled = pd.DataFrame(df_train_scaled, columns=df_train_processed.columns)
df_test_scaled = pd.DataFrame(df_test_scaled, columns=df_test_processed.columns)

# Add unit_number and time_in_cycles back for grouping and reshaping
df_train_scaled[['unit_number', 'time_in_cycles']] = train_unit_time
df_test_scaled[['unit_number', 'time_in_cycles']] = test_unit_time


# Reshape data for CNN-LSTM
def reshape_data(df, sequence_length):
    sequences = []
    for unit_number in df['unit_number'].unique():
        subset = df[df['unit_number'] == unit_number].drop(columns=['unit_number', 'time_in_cycles']).values
        for i in range(len(subset) - sequence_length + 1):
            sequences.append(subset[i:i+sequence_length])
    return np.array(sequences)

# Define sequence length (this is a hyperparameter, can be adjusted)
sequence_length = 50

X_train = reshape_data(df_train_scaled, sequence_length)
X_test = reshape_data(df_test_scaled, sequence_length)


print("Original train shape:", df_train.shape)
print("Processed train shape:", df_train_processed.shape)
print("Scaled train shape:", df_train_scaled.shape)
print("Reshaped train shape:", X_train.shape)

print("\nOriginal test shape:", df_test.shape)
print("Processed test shape:", df_test_processed.shape)
print("Scaled test shape:", df_test_scaled.shape)
print("Reshaped test shape:", X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout, Flatten

# Define input shape
input_shape = (sequence_length, X_train.shape[2])

# Create the hybrid CNN-LSTM model
model = Sequential()

# Add Conv1D layer
model.add(Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=input_shape))

# Add Dropout layer
model.add(Dropout(0.2))

# Add LSTM layer
model.add(LSTM(units=50, return_sequences=False))

# Add another Dropout layer
model.add(Dropout(0.2))

# Add Dense output layer
model.add(Dense(units=1, activation='linear'))

# Print the model summary
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

print("Model compiled successfully.")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("EarlyStopping callback defined.")
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

# ModelCheckpoint callback
model_checkpoint = ModelCheckpoint(
    filepath='best_model.keras',  # .h5 yerine .keras
    monitor='val_loss',
    save_best_only=True
)

print("✅ ModelCheckpoint callback defined to save the best model as 'best_model.keras'.")
```

```python
# Load the RUL data for training
RUL_train_df = pd.read_csv('RUL_FD001.txt', sep='\s+', header=None, names=['RUL'])

# Calculate the RUL for each training data point
# For each unit, the RUL at a given time_in_cycles is the maximum time_in_cycles for that unit
# minus the current time_in_cycles.
max_cycles = df_train.groupby('unit_number')['time_in_cycles'].max().reset_index()
max_cycles.columns = ['unit_number', 'max_time_in_cycles']

df_train_merged = df_train.merge(max_cycles, on='unit_number', how='left')
df_train_merged['RUL'] = df_train_merged['max_time_in_cycles'] - df_train_merged['time_in_cycles']

# Since we are using sequences of length 50, the RUL for a sequence is the RUL of the last data point in the sequence.
# Create the RUL labels for the reshaped training data (X_train)
RUL_train = []
for unit_number in df_train_merged['unit_number'].unique():
    subset = df_train_merged[df_train_merged['unit_number'] == unit_number]
    # The RUL for each sequence is the RUL of the last cycle in that sequence
    for i in range(len(subset) - sequence_length + 1):
        RUL_train.append(subset['RUL'].iloc[i + sequence_length - 1])

RUL_train = np.array(RUL_train)

# Now train the model with early stopping and model checkpoint
history = model.fit(X_train, RUL_train,
                    epochs=100,
                    batch_size=32,
                    validation_split=0.2,
                    callbacks=[early_stopping, model_checkpoint])

print("Model training completed.")
```

```python
# Rastgele Test Verisi Oluşturma
if model is not None:
    print("🎲 Rastgele test verisi oluşturuluyor...")
    
    # Model input shape'ini al
    input_shape = model.input_shape
    print(f"📏 Beklenen input shape: {input_shape}")
    
    # Rastgele test verisi oluştur
    # Shape: (batch_size, timesteps, features)
    if len(input_shape) == 3:  # (None, timesteps, features)
        timesteps = input_shape[1]
        features = input_shape[2]
        
        print(f"📊 Timesteps: {timesteps}, Features: {features}")
        
        # 10 adet rastgele test sample'ı oluştur
        np.random.seed(42)  # Reproducible results için
        n_samples = 10
        
        # Normal dağılımdan rastgele değerler (mean=0, std=1)
        random_test_data = np.random.normal(0, 1, (n_samples, timesteps, features))
        
        print(f"✅ Rastgele test verisi oluşturuldu: {random_test_data.shape}")
        print(f"📈 Veri aralığı: [{random_test_data.min():.2f}, {random_test_data.max():.2f}]")
        
        # İlk sample'ın ilk timestep'ini göster
        print(f"🔍 İlk sample, ilk timestep: {random_test_data[0, 0, :5]}... (ilk 5 özellik)")
        
    else:
        print(f"❌ Beklenmeyen input shape: {input_shape}")
        random_test_data = None
        
else:
    print("❌ Model bulunamadı, rastgele veri oluşturulamadı!")
    random_test_data = None
```

```python
# Model ile Rastgele Tahminler Yapma
if model is not None and random_test_data is not None:
    print("🔮 Model ile rastgele tahminler yapılıyor...")
    
    try:
        # Model ile tahmin yap
        predictions = model.predict(random_test_data, verbose=0)
        
        print(f"✅ Tahminler tamamlandı! {len(predictions)} tahmin yapıldı.")
        print(f"📊 Tahmin shape: {predictions.shape}")
        
        # Tahminleri görüntüle
        print("\n🎯 RASTGELE TAHMİNLER:")
        print("-" * 50)
        for i, pred in enumerate(predictions.flatten()):
            print(f"   Sample {i+1}: RUL = {pred:.2f}")
        
        # İstatistikler
        pred_flat = predictions.flatten()
        print(f"\n📈 TAHMİN İSTATİSTİKLERİ:")
        print(f"   Ortalama: {np.mean(pred_flat):.2f}")
        print(f"   Standart sapma: {np.std(pred_flat):.2f}")
        print(f"   Minimum: {np.min(pred_flat):.2f}")
        print(f"   Maksimum: {np.max(pred_flat):.2f}")
        
    except Exception as e:
        print(f"❌ Tahmin hatası: {e}")
        predictions = None
        
else:
    print("❌ Model veya test verisi mevcut değil!")
    predictions = None
```

```python
# Farklı Senaryolar ile Test
if model is not None:
    print("🧪 Farklı test senaryoları deneniyor...")
    
    # Model input shape'ini al
    timesteps = model.input_shape[1]
    features = model.input_shape[2]
    
    # Senaryo 1: Düşük değerli sensör verileri (erken aşama)
    print("\n📊 SENARYO 1: Erken Aşama Motor (Düşük değerler)")
    early_stage_data = np.random.normal(-1, 0.5, (5, timesteps, features))
    early_predictions = model.predict(early_stage_data, verbose=0)
    
    for i, pred in enumerate(early_predictions.flatten()):
        print(f"   Erken motor {i+1}: RUL = {pred:.2f}")
    
    # Senaryo 2: Yüksek değerli sensör verileri (geç aşama)
    print("\n📊 SENARYO 2: Geç Aşama Motor (Yüksek değerler)")
    late_stage_data = np.random.normal(1, 0.5, (5, timesteps, features))
    late_predictions = model.predict(late_stage_data, verbose=0)
    
    for i, pred in enumerate(late_predictions.flatten()):
        print(f"   Geç motor {i+1}: RUL = {pred:.2f}")
    
    # Senaryo 3: Extremal değerler
    print("\n📊 SENARYO 3: Extremal Değerler")
    extreme_data = np.random.normal(0, 2, (3, timesteps, features))
    extreme_predictions = model.predict(extreme_data, verbose=0)
    
    for i, pred in enumerate(extreme_predictions.flatten()):
        print(f"   Extremal motor {i+1}: RUL = {pred:.2f}")
    
    # Tüm tahminleri karşılaştır
    all_predictions = np.concatenate([
        early_predictions.flatten(),
        late_predictions.flatten(), 
        extreme_predictions.flatten()
    ])
    
    print(f"\n📈 GENEL İSTATİSTİKLER:")
    print(f"   Tüm tahminler - Ortalama: {np.mean(all_predictions):.2f}")
    print(f"   Tüm tahminler - Std: {np.std(all_predictions):.2f}")
    print(f"   Erken aşama - Ortalama: {np.mean(early_predictions):.2f}")
    print(f"   Geç aşama - Ortalama: {np.mean(late_predictions):.2f}")
    
else:
    print("❌ Model bulunamadı!")
```
